```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer.feature_names)
```

```python
df.head()
```

```python
df2 = pd.concat([df, pd.DataFrame(cancer['target'], columns=['Is Benign'])], axis=1)
df2.tail()
```

```python
df2[['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error']]
```

```python
sns.pairplot(df2[['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error','Is Benign']], hue='Is Benign')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df
y = cancer['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(degree=5)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pred))
print()
print(confusion_matrix(y_test, pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parma_grid = {'C':[1,10,100,1000], 'gamma':[1,.1,.01,.001,.0001]}
grid = GridSearchCV(SVC(), param_grid=parma_grid, verbose=2, n_jobs=4)
```

```python
grid.fit(X_train, y_train)
```

```python
pred2 = grid.predict(X_test)
```

```python
pred2 = grid.best_estimator_.predict(X_test)
```

```python
print(classification_report(y_test, pred2))
print()
print(confusion_matrix(y_test, pred2))
```
