```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r*')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Title!")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## OO Method

```python
fig = plt.figure()
axis = fig.add_axes([0.1,0.1,0.8,0.8])
axis.plot(x,y)
axis.set_xlabel("X Label")
axis.set_ylabel("Y Label")
axis.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.3,0.3])

axes1.plot(x,y)
axes1.set_title('larger')
axes2.plot(y,x)
axes2.set_title('smaller')
```

## Figure size
